# Autograd: Gradyan hesaplama

```python
import torch

x = torch.tensor(3.0, requires_grad=True) # requires_grad=True: PyTorch, bu değişkenin türevini hesaplamak için bunu takip et demek
y = x ** 2

print("x =", x)
print("y =", y)
```

```python
y.backward()  # y’nin x’e göre türevini hesapla

print("x.grad =", x.grad)
```

# Matris çarpımı

```python
import torch

A = torch.tensor([[1, 2],
                  [3, 4]])

B = torch.tensor([[5, 6],
                  [7, 8]])

C = torch.matmul(A, B)  # matris çarpımı yapar

print("A =")
print(A)

print("B =")
print(B)

print("A x B =")
print(C)
```

# Reshape

```python
import torch

x = torch.tensor([1, 2, 3, 4, 5, 6])

print("Eski hali:")
print(x)

print("Eski shape:")
print(x.shape)

y = x.reshape(2, 3)  # tensorün boyutunu değiştiri

print("Yeni hali:")
print(y)

print("Yeni shape:")
print(y.shape)
```

# Slicing

```python
import torch

x = torch.tensor([[10, 20, 30],
                  [40, 50, 60],
                  [70, 80, 90]])

print("Orijinal tensor:")
print(x)

print("İlk satır:")
print(x[0])                                #tensorün içinden istediğin parçayı alma

print("İkinci satırın üçüncü elemanı:")
print(x[1, 2])

print("İlk iki satır ve ilk iki sütun:")
print(x[:2, :2])
```
