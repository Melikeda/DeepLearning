```python
import torch
import torch.nn as nn
```

```python
class MyModel(nn.Module):
    def __init__(self):
        super().__init__()
        
        self.fc1 = nn.Linear(3, 5)
        self.fc2 = nn.Linear(5, 1)

    def forward(self, x):
        print("Gelen input:", x)
        
        x = self.fc1(x)
        print("fc1 sonrası:", x)
        
        x = self.fc2(x)
        print("fc2 sonrası:", x)
        
        return x
```

```python
model = MyModel()
print(model)
```

```python
x = torch.tensor([[1.0, 2.0, 3.0]])
print(x)
```

```python
output = model(x)
print(output)
```

### class MyModel(nn.Module):

Burada kendi modelini oluşturuyorsun.

MyModel = modelin adı
(nn.Module) = PyTorch model şablonunu kullanıyorsun

### def __init__(self):

Bu bölüm model kurulduğu anda çalışır.

Burada: katmanları tanımlarsın, modelin yapısını kurarsın

### super().__init__()

Bu satır çok önemli.

Bu, nn.Module’ün içindeki hazır sistemi başlatır.

Bunu koymazsan PyTorch modelini doğru tanımaz.

### self.fc1 = nn.Linear(3, 5)

Bu ilk katman.

Anlamı:

girişte 3 sayı alacak
çıkışta 5 sayı üretecek

### self.fc2 = nn.Linear(5, 1)

Bu ikinci katman.

Anlamı:

5 sayı alır
1 sayı çıkarır

### forward kısmı 

Bu bölüm şunu söyler:

“Gelen veri modelin içinde hangi sırayla ilerleyecek?”

Burada akış şu:

veri gelir: x
fc1 katmanına girer
çıkan sonuç fc2 katmanına gider
en son çıktı döner

Yani forward:
verinin yol haritası
