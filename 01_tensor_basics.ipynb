# Temel İşlemler

```python
import torch

x = torch.tensor([1, 2, 3])
print(x)
```

### NumPy karşılığı:

```python
import numpy as np

x = np.array([1, 2, 3])
```

## Shape

```python
x = torch.tensor([[1,2,3],[4,5,6]])
print(x.shape)
```

### NumPy karşılığı:

```python
x = np.array([[1,2,3],[4,5,6]])
print(x.shape)
```

## dtype

```python
x = torch.tensor([1,2,3])
print(x.dtype)
```

```python
x = torch.tensor([1,2,3], dtype=torch.float32)
print(x.dtype)
```

### NumPy karşılığı:

```python
x = np.array([1,2,3], dtype=np.float32)
```
